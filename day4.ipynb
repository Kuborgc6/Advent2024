```python
def checkAdjacent(i, j, lines, letter):
  if i >= 0 and i <= len(lines) - 1 and j >= 0 and j <= len(lines[i]) - 1:
    if lines[i][j] == letter:
      return True
  return False
```

```python
result = 0
wayi = [-1,0,1]
wayj = [-1,0,1]
for veci in wayi:
  for vecj in wayj:
    for i in range(len(lines)):
# i = 9
# if True:
  # print(lines[i])
      for j in range(len(lines[i])):
    # i = 0
    # j = 4
        # print(lines[i][j])
        # print(lines[i])
        # print(lines[1][5])
        if checkAdjacent(i, j, lines, 'X'):
          # print('test' + lines[i+wayi][j+wayj])
          if checkAdjacent(i+veci, j+vecj, lines, 'M'):
            # print(lines[i+2*wayi][j+2*wayj])
            if checkAdjacent(i+2*veci, j+2*vecj, lines, 'A'):
              # print(lines[i+3*wayi][j+3*wayj])
              if checkAdjacent(i+3*veci, j+3*vecj, lines, 'S'):
                # result += [lines[i][j]+lines[i+wayi][j+wayj]+lines[i+2*wayi][j+2*wayj]+lines[i+3*wayi][j+3*wayj]]
                result += 1

print(result)
```

```python
result = 0
wayi = [-1,1]
wayj = [-1,1]

for i in range(len(lines)):
    for j in range(len(lines[i])):
      break_flag = False
      for veci in wayi:
        for vecj in wayj:
          if checkAdjacent(i, j, lines, 'A'):
            if (checkAdjacent(i+veci, j+vecj, lines, 'M') & checkAdjacent(i-veci, j-vecj, lines, 'S')
            &
                checkAdjacent(i-veci, j+vecj, lines, 'M') & checkAdjacent(i+veci, j-vecj, lines, 'S')):
              result += 1
              break_flag = True
              break
            if (checkAdjacent(i+veci, j+vecj, lines, 'M') & checkAdjacent(i-veci, j-vecj, lines, 'S')
            &
                checkAdjacent(i+veci, j-vecj, lines, 'M') & checkAdjacent(i-veci, j+vecj, lines, 'S')):
              result += 1
              break_flag = True
              break
        if break_flag:
          break

print(result)

```
