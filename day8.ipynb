```python
input = '/content/sample_data/input.txt'
# input = '/content/sample_data/input_test.txt'
```

```python
file = open(input, 'r')
lines_clean = [list(line.split()[0]) for line in file]
lines = [line.copy() for line in lines_clean]
```

```python
# lines
```

```python
antenas_dict = dict()
for i in range(len(lines)):
  for j in range(len(lines[i])):
    if lines[i][j] != '.':
      current = lines[i][j]
      antenas_dict[current] = [[i,j]]
      lines[i][j] = '.'
      for k in range(len(lines)):
        for l in range(len(lines[k])):
          if lines[k][l] == current:
            lines[k][l] = '.'
            antenas_dict[current].append([k,l])
print(antenas_dict)
```

```python
result = set()
for key in antenas_dict:
  if len(antenas_dict[key]) > 1:
    for i in range(len(antenas_dict[key])):
      for j in range(i+1, len(antenas_dict[key])):
        vector = [antenas_dict[key][i][0] - antenas_dict[key][j][0], antenas_dict[key][i][1] - antenas_dict[key][j][1]]
        antinodes_first = [antenas_dict[key][i][0] + vector[0], antenas_dict[key][i][1] + vector[1]]
        antinodes_second = [antenas_dict[key][j][0] - vector[0], antenas_dict[key][j][1] - vector[1]]
        if antinodes_first[0] >= 0 and antinodes_first[0] < len(lines) and antinodes_first[1] >= 0 and antinodes_first[1] < len(lines[0]):
          result.add(tuple(antinodes_first))
        if antinodes_second[0] >= 0 and antinodes_second[0] < len(lines) and antinodes_second[1] >= 0 and antinodes_second[1] < len(lines[0]):
          result.add(tuple(antinodes_second))
print(len(result))
```

```python
result = set()
for key in antenas_dict:
  if len(antenas_dict[key]) > 1:
    for i in range(len(antenas_dict[key])):
      for j in range(i+1, len(antenas_dict[key])):
        vector = [antenas_dict[key][i][0] - antenas_dict[key][j][0], antenas_dict[key][i][1] - antenas_dict[key][j][1]]
        antinodes_first = [antenas_dict[key][i][0] + vector[0], antenas_dict[key][i][1] + vector[1]]
        antinodes_second = [antenas_dict[key][j][0] - vector[0], antenas_dict[key][j][1] - vector[1]]
        while antinodes_first[0] >= 0 and antinodes_first[0] < len(lines) and antinodes_first[1] >= 0 and antinodes_first[1] < len(lines[0]):
          result.add(tuple(antinodes_first))
          antinodes_first[0] += vector[0]
          antinodes_first[1] += vector[1]
        while antinodes_second[0] >= 0 and antinodes_second[0] < len(lines) and antinodes_second[1] >= 0 and antinodes_second[1] < len(lines[0]):
          result.add(tuple(antinodes_second))
          antinodes_second[0] -= vector[0]
          antinodes_second[1] -= vector[1]
print(len(result))
```

```python
lines = [line.copy() for line in lines_clean]
```

```python
for i in result:
  if lines[i[0]][i[1]] == '.':
    lines[i[0]][i[1]] = '#'
```

```python
result_qt = 0
for line in lines:
  for char in line:
    if char != '.':
      result_qt += 1
print(result_qt)
```
